```python
import random
import pandas as pd

random.seed(2022)  # seed

### Generace dat ###
y = []
for x in range(0, 100):
    y.append(random.randint(0, x))

data = pd.DataFrame({"x": range(0, 100), "y": y})
y.reverse()
data["y_2"] = y

y = []
for x in range(0, 100):
    y.append(random.randint(max(x,100-x), 100))
data["y_3"] = y
```

```python
from matplotlib import pyplot as plt

ax = plt.figure(figsize=(10, 4), dpi=100).add_subplot()
ax.plot(data["x"], data["y"], linestyle="none", marker="o")
ax.plot(data["x"], data["y_2"], linestyle="none", marker="x")
ax.plot(data["x"], data["y_3"], linestyle="none", marker="^")
ax.legend(["Data A", "Data B", "Data C"])
plt.show()
```
