```python
import numpy as np
import pandas as pd
import random
import math

random.seed(2022) #seed

### Generace dat ###
x_data = np.linspace(0, math.pi*4, 1000)
y_data_a = []
y_data_b = []
for x in x_data:
    y_data_a.append(math.sin(x)+math.sin(x*3)+math.sin(x*5)+math.sin(x*7)+math.sin(x*10))
    y_data_b.append(sum(y_data_a[-250:-1])/250)
    
data = pd.DataFrame({"x":x_data,"data_a":y_data_a,"data_b":y_data_b})
```

## Graf bez popisků a kontextu

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(10, 4), dpi=100)
ax1 = fig.add_subplot()

ax1.plot(data["x"],data["data_a"])
ax1.plot(data["x"],data["data_b"])
ax1.axis("off")
ax1.set_xbound(0,math.pi*4)
```

## Graf s popisky

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(10, 4), dpi=100)
ax1 = fig.add_subplot()

ax1.plot(data["x"],data["data_a"])
ax1.plot(data["x"],data["data_b"])
ax1.set_xlabel("Název osy X")
ax1.set_ylabel("Název osy Y")
ax1.set_title("Nadpis grafu")
ax1.legend(["Data A", "Data B"], title="Nadpis legendy")
ax1.set_xbound(0,math.pi*4)
ax1.grid(True, linestyle=":")
```
